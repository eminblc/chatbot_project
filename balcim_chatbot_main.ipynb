```python
from pathlib import Path

def srt_to_txt_raw(srt_path, encoding="cp1254"):
    content = Path(srt_path).read_text(encoding=encoding, errors="ignore")
    out_path = Path(srt_path).with_suffix(".txt")
    out_path.write_text(content, encoding="utf-8")
root = r"C:\Users\eminn\Documents\mat409\balcim_chatbot_projesi\data"

print("Converting .srt files to .txt...")
print(len(list(Path(root).rglob("*.srt"))), "files found.")
for srt in Path(root).rglob("*.srt"):
    srt_to_txt_raw(srt)
    print("OK:", srt.name)
```

```python
import pysubs2

# def srt_txt_to_plain_text(fp, encoding="cp1254"):
#     subs = pysubs2.load(fp, encoding=encoding)  # SRT okur
#     # event.plaintext: tagsiz metin (b/i gibi tag’ler gider)
#     lines = [ev.plaintext.strip() for ev in subs if ev.plaintext.strip()]
#     return "\n".join(lines)
# # def srt_txt_to_plain_text(fp, encoding="cp1254"):
# #     subs = pysubs2.load(fp, encoding=encoding, format_="srt")  # <-- BURASI
# #     lines = [ev.plaintext.strip() for ev in subs if ev.plaintext.strip()]
# #     return "\n".join(lines)


def srt_txt_to_plain_text(fp):
    # Windows altyazılarında en sık görülenler
    encodings = ["cp1254", "utf-8", "iso-8859-9", "cp1252"]

    last_err = None
    for enc in encodings:
        try:
            subs = pysubs2.load(
                fp,
                encoding=enc,
                errors="replace",   # <-- patlamasın, bozuk karakteri � yapar
                format_="srt"       # <-- autodetect yok, SRT diye zorla
            )
            lines = [ev.plaintext.strip() for ev in subs if ev.plaintext.strip()]
            return "\n".join(lines)
        except Exception as e:
            last_err = e
            continue

    raise last_err
```

```python
from langchain_core.documents import Document

clean_data = []
for fp in txt_files:
    # plain = srt_txt_to_plain_text(fp, encoding="cp1254")
    plain = srt_txt_to_plain_text(fp)
    clean_data.append(Document(page_content=plain, metadata={"source": fp}))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
    is_separator_regex=False,
)
```

```python
docs = []
for doc in clean_data:
    chunks = text_splitter.split_documents([doc])
    docs.extend(chunks)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name="stranger_things_season_1",
)
```

```python
from langchain_google_genai import GoogleGenerativeAI
llm = GoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.2, # Daha tutarlı cevaplar için düşük sıcaklık,
    max_tokens=500, # Maksimum token sayısı
)
```

```python
system_prompt = (
     """You are a question-answering assistant specialized in the TV series "Stranger Things".

You must follow these rules strictly:

1) Use ONLY the information provided in the given context.
2) If the answer is NOT explicitly contained in the context, respond exactly with:
   "Bilmiyorum."
3) Do NOT make up, infer, guess, or assume any information that is not present in the context.
4) Do NOT answer questions that are unrelated to the provided context.
5) If the context does not specify who said a line, do NOT guess the speaker.
   In such cases, state that the speaker is not specified.
6) Answer ONLY the question that is asked. Do not add extra explanations.
7) All answers MUST be written in Turkish.

Special rule:
- If the question is exactly "Stranger Things nedir?", respond with:
  "Stranger Things, Duffer Kardeşler tarafından yaratılan ve Netflix'te yayınlanan popüler bir bilim kurgu ve korku dizisidir. 1980'lerde küçük bir kasabada geçen dizi, kaybolan bir çocuğun etrafında dönen gizemli olayları ve doğaüstü güçleri konu alır."
"""
    
   

     "\n\n"
    "{context}"
)
```

```python
import langchain_core.prompts as chatprompts
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
```

```python
prompt = chatprompts.ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("user", "{input}"),
    ]
)
```

```python
question_answering_chain = create_stuff_documents_chain(
    llm=llm,
    prompt=prompt
)   
```

```python
rag_chain = create_retrieval_chain(
    retriever,
    question_answering_chain
)
```

```python
response = rag_chain.invoke({"input": "Dizide çocukların oynadığı Dungeons & Dragons oyununun önemi nedir?"})
print(response)
print(response["answer"])
    
```
